# 5) Membersihkan data

```python
# Dalam dataset ini, beberapa fitur memiliki hampir 90% data yang bernilai null.
# Oleh karena itu, kita perlu menghapus fitur-fitur tersebut menggunakan fungsi drop.

# Mendefinisikan kolom yang akan dihapus
columns_to_drop = ['ca', 'slope','thal']

# Menghapus kolom yang telah ditentukan dari DataFrame
df_selected = df_selected.drop(columns_to_drop, axis=1)

# Menghitung jumlah nilai yang hilang dalam setiap kolom dalam DataFrame
df_selected.isnull().sum()
```

```python
# Karena masih ada nilai null di beberapa kolom fitur, kita perlu mengisi nilai null tersebut
# dengan nilai rata-rata di setiap kolomnya.

# Menghapus nilai null pada kolom tertentu menggunakan fungsi dropna()
# Misalnya, meanTBPS = df_selected['trestbps'].dropna() menghapus semua baris yang memiliki nilai null pada kolom ‘trestbps’ dan menyimpan hasilnya dalam variabel meanTBPS.
meanTBPS = df_selected['trestbps'].dropna()
meanChol = df_selected['chol'].dropna()
meanfbs = df_selected['fbs'].dropna()
meanRestCG = df_selected['restecg'].dropna()
meanthalach = df_selected['thalach'].dropna()
meanexang = df_selected['exang'].dropna()

# Mengubah tipe data menjadi float menggunakan fungsi astype(float)
# Menghitung rata-rata dari setiap kolom menggunakan fungsi mean()
meanTBPS = meanTBPS.astype(float)
meanChol = meanChol.astype(float)
meanfbs = meanfbs.astype(float)
meanthalach = meanthalach.astype(float)
meanexang = meanexang.astype(float)
meanRestCG = meanRestCG.astype(float)

# Membulatkan hasil rata-rata ke bilangan bulat terdekat menggunakan fungsi round().
# Misalnya, meanTBPS = round(meanTBPS.mean()) membulatkan rata-rata dari semua nilai dalam variabel meanTBPS ke bilangan bulat terdekat.
meanTBPS = round(meanTBPS.mean())
meanChol = round(meanChol.mean())
meanfbs = round(meanfbs.mean())
meanthalach = round(meanthalach.mean())
meanexang = round(meanexang.mean())
meanRestCG = round(meanRestCG.mean())
```

```python
# Membuat dictionary yang berisi pasangan key-value, di mana key adalah nama kolom
# dan value adalah nilai rata-rata yang telah dihitung sebelumnya untuk setiap kolom tersebut.
fill_values = {'trestbps': meanTBPS, 'chol': meanChol, 'fbs': meanfbs,
               'thalach':meanthalach,'exang':meanexang,'restecg':meanRestCG}

# Mengisi nilai null dalam DataFrame df_selected dengan nilai rata-rata yang sesuai
# dari dictionary fill_values. Hasilnya disimpan dalam DataFrame baru bernama dfClean.
dfClean = df_selected.fillna(value=fill_values)

# Menampilkan ringkasan singkat tentang DataFrame dfClean
dfClean.info()
```

* Grafik yang ditampilkan sebelumnya menunjukkan bahwa distribusi target dalam data kita tidak seimbang. Oleh karena itu, kita perlu melakukan penyeimbangan. Ada dua metode umum untuk melakukan ini: oversampling dan undersampling.

* Oversampling biasanya digunakan ketika jumlah data kita relatif sedikit

* Undersampling digunakan ketika kita memiliki jumlah data yang sangat besar.

* Dalam kasus kita, karena jumlah data kita tidak terlalu banyak, kita akan menggunakan metode oversampling.

* Salah satu teknik yang bisa kita gunakan untuk melakukan oversampling adalah SMOTE (Synthetic Minority Over-sampling Technique).

* Metode ini akan membantu kita dalam menyeimbangkan distribusi target dalam data kita

```python
# Mengimpor library yang diperlukan
from imblearn.over_sampling import SMOTE

# Melakukan oversampling menggunakan SMOTE
smote = SMOTE(random_state=42)
X_smote_resampled, y_smote_resampled = smote.fit_resample(X, y)

# Membuat figure dan subplot untuk plot
plt.figure(figsize=(12, 4))
new_df1 = pd.DataFrame(data=y)

# Menampilkan jumlah target sebelum oversampling dengan SMOTE
plt.subplot(1, 2, 1)
new_df1.value_counts().plot(kind='bar',figsize=(10,6),color=['green','blue','red','yellow'])
plt.title("target before over sampling with SMOTE ")
plt.xticks(rotation=0);

# Menampilkan jumlah target setelah oversampling dengan SMOTE
plt.subplot(1, 2, 2)
new_df2 = pd.DataFrame(data=y_smote_resampled)
new_df2.value_counts().plot(kind='bar',figsize=(10,6),color=['green','blue','red','yellow'])
plt.title("target after over sampling with SMOTE")
plt.xticks(rotation=0);

# Menyesuaikan layout dan menampilkan plot
plt.tight_layout()
plt.show()
```

# 7) Model

Ini adalah fungsi yang dirancang untuk menghasilkan nilai akurasi dan rata-rata dari recall, f1, dan precision score untuk setiap model. Fungsi ini akan dijalankan pada setiap model. Pembuatan fungsi ini bersifat pilihan, bukan keharusan.

```python
from sklearn.metrics import accuracy_score,recall_score,f1_score,precision_score,roc_auc_score,confusion_matrix,precision_score
```

```python
# Mendefinisikan fungsi 'evaluation' dengan dua parameter: Y_test dan Y_pred.
def evaluation(Y_test,Y_pred):
  # Menghitung akurasi prediksi dengan membandingkan Y_test (nilai sebenarnya) dan Y_pred (nilai yang diprediksi).
  acc = accuracy_score(Y_test,Y_pred)
  # Menghitung recall (sensitivitas) dengan membandingkan Y_test dan Y_pred. Parameter 'average' diatur ke 'weighted' untuk menghitung rata-rata tertimbang dari recall untuk setiap kelas.
  rcl = recall_score(Y_test,Y_pred,average = 'weighted')
  # Menghitung F1 score (rata-rata harmonik dari presisi dan recall) dengan membandingkan Y_test dan Y_pred. Parameter 'average' diatur ke 'weighted' untuk menghitung rata-rata tertimbang dari F1 score untuk setiap kelas.
  f1 = f1_score(Y_test,Y_pred,average = 'weighted')
  # Menghitung presisi (ketepatan) dengan membandingkan Y_test dan Y_pred. Parameter 'average' diatur ke 'weighted' untuk menghitung rata-rata tertimbang dari presisi untuk setiap kelas.
  ps = precision_score(Y_test,Y_pred,average = 'weighted')
  # Membuat dictionary untuk menyimpan nilai akurasi dan membulatkannya hingga 3 angka desimal.
  metric_dict={'accuracy': round(acc,3),
    # Menyimpan nilai recall dan membulatkannya hingga 3 angka desimal.
    'recall': round(rcl,3),
    # Menyimpan nilai F1 score dan membulatkannya hingga 3 angka desimal.
    'F1 score': round(f1,3),
    # Menyimpan nilai presisi dan membulatkannya hingga 3 angka desimal.
    'Precision score': round(ps,3)
  }

  # Mencetak dictionary yang berisi metrik evaluasi.
  return print(metric_dict)
```

Algoritma ini telah terbukti sangat efektif dalam berbagai kompetisi pembelajaran mesin dan memiliki dukungan untuk berbagai fungsi kerugian dan metrik evaluasi, penanganan data yang hilang, dan regularisasi untuk mencegah overfitting.

* accuracy_score: Fungsi ini menghitung akurasi klasifikasi, yaitu proporsi sampel yang diklasifikasikan dengan benar.

* classification_report: Fungsi ini membangun laporan teks yang menunjukkan metrik klasifikasi utama, seperti presisi, recall, dan F1-score.

Dalam penelitian ini, saya menggunakan berbagai teknik penanganan data, termasuk random over sampling SMOTE untuk mengatasi data yang tidak seimbang, RandomSearchCV untuk penyetelan parameter, dan normalisasi data.

* Hasil penelitian menunjukkan bahwa klasifikasi dengan Random Over Sampling SMOTE pada model KNN menghasilkan akurasi 75.4%, model Random Forest menghasilkan akurasi 92%, dan model XGBoost menghasilkan akurasi 90.4%. Namun, ketika data telah dinormalisasi dan ditangani dengan Random Over Sampling SMOTE, model KNN menghasilkan akurasi yang meningkat menjadi 86.1%, sementara model Random Forest dan XGBoost tetap stabil dengan akurasi 92% dan 90.4% masing-masing.

* Lebih lanjut, ketika data telah ditangani dengan tunning RandomSearchCV, normalisasi, dan Random Over Sampling SMOTE, model KNN menghasilkan akurasi tertinggi yaitu 93%, diikuti oleh model XGBoost dengan akurasi 92%, dan model Random Forest dengan akurasi 87.7%.

* Dari hasil penelitian ini, dapat disimpulkan bahwa penanganan data yang optimal untuk mengatasi ketidakseimbangan data adalah dengan menggunakan metode random Oversampling SMOTE yang dilengkapi dengan penyetelan parameter menggunakan RandomSearchCV dan normalisasi data. Hal ini memberikan hasil yang signifikan dalam meningkatkan akurasi model klasifikasi, khususnya pada model KNN dan XGBoost. Namun, hal ini tidak berlaku untuk model Random Forest yang justru mengalami penurunan akurasi yang signifikan.

* Secara keseluruhan, penanganan ketidakseimbangan data dengan menggunakan penyetelan parameter, normalisasi, dan oversampling dapat memberikan dampak signifikan terhadap performa model klasifikasi. Pemilihan model dan parameter yang optimal dapat meningkatkan akurasi dan kinerja model secara keseluruhan.
